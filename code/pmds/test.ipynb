A square matrix A is diagonalizable if it can be expressed in the form:
A = PDP^−1

- P is a matrix whose columns are the eigenvectors of A.

- D is a diagonal matrix whose diagonal entries are the eigenvalues of A.

```python
def can_be_diagonalized(A):
    eigenvalues, eigenvectors = np.linalg.eig(A)
    print("eigenvectors:\n",eigenvectors)
    D = np.diag(eigenvalues)
    print("eigen_vals:\n", D)
    inverse_P = np.linalg.inv(eigenvectors)
    print("inverse_P:\n", inverse_P)
    return np.allclose(A, eigenvectors@D@inverse_P)

print(can_be_diagonalized(D))
```

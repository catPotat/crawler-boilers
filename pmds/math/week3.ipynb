```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_true = np.array([[10], [5]])
X = np.random.uniform(0, 100, 100).reshape(100, 1)
X = np.concatenate((np.ones((100, 1)), X), axis=1)
y = X @ theta_true + np.random.normal(0, 10, (100, 1))
```

```python
plt.plot(X[:, 1], y, "o")
plt.xlabel("Input X")
plt.ylabel("y=Output y")
plt.title("Synthetic data for testing linear regression")
plt.show()
```

```python
# A.T@A@w = A.T@y

theta_star = np.linalg.inv(X.T @ X) @ X.T @ y
theta_star
```

```python
x1 = 0
y1 = theta_star[0]
plt.plot(X[:, 1], y, "o")

x2 = 100
y2 = theta_star[0] + theta_star[1] * x2
plt.plot((x1, x2), (y1, y2), "b-", "Estimated regression line")

x1 = 0
y1 = theta_true[0]
x2 = 100
y2 = theta_true[0] + theta_true[1] * x2
plt.plot((x1, x2), (y1, y2), "r-", "True regression line")
```

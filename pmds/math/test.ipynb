A square matrix A is diagonalizable if it can be expressed in the form:
A = PDP^−1

- P is a matrix whose columns are the eigenvectors of A.

- D is a diagonal matrix whose diagonal entries are the eigenvalues of A.

```python
def can_be_diagonalized(A):
    eigenvalues, eigenvectors = np.linalg.eig(A)
    print("eigenvectors:\n",eigenvectors)
    D = np.diag(eigenvalues)
    print("eigen_vals:\n", D)
    inverse_P = np.linalg.inv(eigenvectors)
    print("inverse_P:\n", inverse_P)
    return np.allclose(A, eigenvectors@D@inverse_P)

print(can_be_diagonalized(D))
```

```python
import numpy as np

# Define matrix E
E = np.array([[4, 1, -2], [2, 3, 1], [-1, 0, 2]])
print("Matrix E:\n", E)

#Compute the eigenvalues and eigenvectors of E using NumPy.
eigenvalues, eigenvectors = np.linalg.eig(E)

print("Eigenvalues of E:\n", eigenvalues)
print("\nEigenvectors of E:\n", eigenvectors)

#Matrix P using eigenvectors
P = eigenvectors
print("\nMatrix P:\n", P)

#Diagonal Matrix D using Eigenvalues
D = np.diag(eigenvalues)
print("\nDiagonal Matrix D:\n", D)

#Verify if E = PDP^-1
E_check = P @ D @ np.linalg.inv(P)
is_diagonalizable = np.allclose(E, E_check)

print("Is E diagonalizable?", is_diagonalizable)

from scipy.linalg import expm

# Calculate exp(D)
exp_D = np.diag(np.exp(eigenvalues))

# Calculate exp(E) using P exp(D) P^-1
exp_E_manual = P @ exp_D @ np.linalg.inv(P)

# Calculate exp(E) using built-in function
exp_E_builtin = expm(E)

# Compare the results
print("exp(E) calculated manually:\n", exp_E_manual)
print("exp(E) calculated using scipy.linalg.expm:\n", exp_E_builtin)

# Verify if both results match
is_match = np.allclose(exp_E_manual, exp_E_builtin)
print("Manually matched scipy.linalg.expm:\n ", is_match)
```
